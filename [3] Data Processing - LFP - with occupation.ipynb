```
campers.model <- ipv[, c("X", "V007", "V012", "V024", "V106", "V107", "V149", "V190", "V190A", "V501", "V701", "V702", "V704", "V705",
"V716","V717", "V729", "V730", "D104", "D106", "D107", "D108")]

colnames(campers.model) <- c("RespondentID", "Year", "Age", "Region", "Educ.Level", "Educ.Years", "Educ.Attained", "Wealth.Index", "Wealth.Index.Rural", "Marital", "Husb.Educ.Level", 
"Husb.Educ.Years", "Husb.Employment", "Husb.Occupation", "Employment", "Occupation", "Husb.Educ.Attained", "Husb.Age", "Emotional.IPV", "Less.Physic.IPV", 
"More.Physic.IPV", "Sexual.IPV")
```

```
View(campers.model)
```

```
# Binary Indicator for Employment for Years 2008, 2013, 2017

campers.model.2 <- campers.model %>%
    mutate(Husb.Employment.Bin = case_when(
        Year == 2008 & Husb.Employment == 21 | Husb.Employment == 22 | 
            Husb.Employment == 23 | Husb.Employment == 94 ~ 0,
        Year == 2008 & Husb.Employment != 21 & Husb.Employment != 22 & 
            Husb.Employment != 23 & Husb.Employment != 94 & Husb.Employment != 999 ~ 1,
        Year == 2008 & Husb.Employment == 999 ~ 2,
        Year == 2013 & Husb.Employment == 21 | Husb.Employment == 22 | 
            Husb.Employment == 23 | Husb.Employment == 94 ~ 0,
        Year == 2013 & Husb.Employment != 21 & Husb.Employment != 22 & 
            Husb.Employment != 23 & Husb.Employment != 94 & Husb.Employment != 998 & Husb.Employment != 999 ~ 1,
        Year == 2013 & Husb.Employment == 998 | Husb.Employment == 999 ~ 2,
        Year == 2017 & Husb.Employment == 0 |Husb.Employment == 21 | Husb.Employment == 22 | 
            Husb.Employment == 23 | Husb.Employment == 94 ~ 0,
        Year == 2017 & Husb.Employment !=0 & Husb.Employment != 21 & Husb.Employment != 22 & 
            Husb.Employment != 23 & Husb.Employment != 94 & Husb.Employment != 9998 & Husb.Employment != 9999 ~ 1,
        Year == 2017 & Husb.Employment == 9998 | Husb.Employment == 9999 ~ 2,
        
    ))

campers.model.3 <- campers.model.2 %>%
    mutate(Employment.Bin = case_when(
        Year == 2008 & Employment == 0 | Employment == 21 | Employment == 22 | 
            Employment == 23 | Employment == 94 ~ 0,
        Year == 2008 & Employment != 0 & Employment != 21 & Employment != 22 & 
            Employment != 23 & Employment != 94 & Employment != 999 ~ 1,
        Year == 2008 & Employment == 999 ~ 2,
        Year == 2013 & Employment == 0 | Employment == 21 | Employment == 22 | 
            Employment == 23 | Employment == 94 ~ 0,
        Year == 2013 & Employment !=0 & Employment != 21 & Employment != 22 & 
            Employment != 23 & Employment != 94 & Employment != 998 & Employment != 999 ~ 1,
        Year == 2013 & Employment == 998 | Employment == 999 ~ 2,
        Year == 2017 & Employment == 0 | Employment == 21 | Employment == 22 | 
            Employment == 23 | Employment == 94 ~ 0,
        Year == 2017 & Employment !=0 & Employment != 21 & Employment != 22 & 
            Employment != 23 & Employment != 94 & Employment != 9998 & Employment != 9999 ~ 1,
        Year == 2017 & Employment == 9998 | Employment == 9999 ~ 2,
    ))

campers.model.final <- campers.model.3[campers.model.3$Employment.Bin < 2 & campers.model.3$Husb.Employment.Bin <2, ]


# Accounting for reduced sample size
nrow(ipv)
nrow(campers.model.final)
```

```
# Adding variable for Labor Participaton Gap

campers.model.lpg <- campers.model.final %>%
    mutate(LPG = case_when(
        Husb.Employment.Bin == 0 & Employment.Bin == 0 ~ "Both unemployed",
        Husb.Employment.Bin == 0 & Employment.Bin == 1 ~ "Female only Employed",
        Husb.Employment.Bin == 1 & Employment.Bin == 0 ~ "Male only Employed",
        Husb.Employment.Bin == 1 & Employment.Bin == 1 ~ "Both employed"
    ))
```

```
# Adding variable for Physical IPV

campers.model.physical <- campers.model.lpg %>%
    mutate(Physical.IPV = case_when(
        Less.Physic.IPV > 0 | More.Physic.IPV > 0 ~ 1,
        Less.Physic.IPV == 0 & More.Physic.IPV == 0 ~ 0
    ))
```

```
# Adding binary indicator variable for Any IPV

campers.model.any.ipv <- campers.model.physical %>%
    mutate(ANY.IPV = case_when(
        Physical.IPV > 0 | Sexual.IPV > 0 | Emotional.IPV > 0 ~ 1,
        Physical.IPV == 0 & Sexual.IPV == 0 & Emotional.IPV == 0 ~ 0,
    ))
```

```
# Adding variable for Total Counts of IPV

campers.model.counts.ipv <- campers.model.any.ipv %>%
    mutate(Counts.IPV = case_when(
        Physical.IPV >= 0 & Sexual.IPV >= 0 & Emotional.IPV >= 0 ~ Physical.IPV + Sexual.IPV + Emotional.IPV
    ))
```

```
View(campers.model.counts.ipv)
colnames(campers.model.counts.ipv)
```

```
# Removing rows that are NA since they didn't give consent about questions on IPV

campers.ipv.for.analysis <- campers.model.counts.ipv[complete.cases(campers.model.counts.ipv[, c("ANY.IPV")]),]
```

```
View(campers.ipv.for.analysis)
```
